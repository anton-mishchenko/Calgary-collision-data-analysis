```python
columns = ["ID","LOCAL_DATE","STATION_NAME","MEAN_TEMPERATURE","TOTAL_PRECIPITATION","TOTAL_RAIN","TOTAL_SNOW","MAX_REL_HUMIDITY","MIN_REL_HUMIDITY","SPEED_MAX_GUST"]
weather = weather[columns]
weather
```

```python
weather.LOCAL_DATE = weather.LOCAL_DATE.str.split(' ').str[0]
weather
```

```python
# combining data from 3 stations
columns = ["ID","LOCAL_DATE","STATION_NAME","MEAN_TEMPERATURE","TOTAL_PRECIPITATION","TOTAL_RAIN","TOTAL_SNOW","MAX_REL_HUMIDITY","MIN_REL_HUMIDITY","SPEED_MAX_GUST"]

aggregation_functions = {
    'LOCAL_DATE': 'first',
    'MEAN_TEMPERATURE': 'mean',
    'TOTAL_PRECIPITATION': 'mean',
    'TOTAL_RAIN': 'mean',
    'TOTAL_SNOW': 'mean',
    'MAX_REL_HUMIDITY': 'mean',
    'MIN_REL_HUMIDITY': 'mean',
    'SPEED_MAX_GUST': 'mean'

}
combined_weather = weather.groupby(weather['LOCAL_DATE']).aggregate(aggregation_functions)
combined_weather
```

```python
combined_weather.reset_index(drop=True, inplace=True)
combined_weather
```

```python
incidents = pd.read_csv("calgary_incidents.csv")
```

```python
incident_columns = ["id","START_DT","Count"]
incidents = incidents[incident_columns]
incidents
```

```python
incidents.START_DT = incidents.START_DT.str.split(' ').str[0]
incidents
```

```python
aggregation_functions = {
    'START_DT': 'first',
    'Count': 'sum'
}

combined_incidents = incidents.groupby(incidents['START_DT']).aggregate(aggregation_functions)
combined_incidents
```

```python
combined_incidents.reset_index(drop=True, inplace=True)
combined_incidents
```

```python
new_cols_dict = {
    'START_DT': 'LOCAL_DATE',
    'Count': 'COUNT'
}

combined_incidents.rename(new_cols_dict, axis=1, inplace=True)
combined_incidents
```

```python
newdf = combined_weather.merge(combined_incidents, on="LOCAL_DATE", how="inner")
newdf
```
